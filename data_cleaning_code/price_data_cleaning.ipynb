```python
import pandas as pd
import numpy as np
from nemosis import static_table, dynamic_data_compiler
import plotly.express as px
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# Specify where we will be caching the raw AEMO data.
raw_data_cache = '/Volumes/T7/NEMO_data'
```

RRP (Regional Reference Price): Indicates that this price is region-specific and represents the market-clearing price for the lower regulation reserve.

LOWERREGRRP stands for Lower Regulation Reserve Regional Reference Price. It represents the price for lower regulation ancillary services in a specific region within the electricity market.

```python
start_time = '2003/01/01 00:00:00'
end_time = '2024/01/01 00:05:00'
table = 'DISPATCHPRICE'

price_data = dynamic_data_compiler(start_time, end_time, table, raw_data_cache)
```

```python
price_data.head()
```

```python
#Create a price time series for a single day
specific_day = '2020-09-27'
filtered_data = price_data[price_data["SETTLEMENTDATE"].dt.date == pd.to_datetime(specific_day).date()]
pivoted_data = filtered_data.pivot(index="SETTLEMENTDATE", columns="REGIONID", values="RRP")

# Plot the prices for the specified day
plt.figure(figsize=(10, 6))
for region in pivoted_data.columns:
    plt.plot(pivoted_data.index, pivoted_data[region], label=region)

# Note there are 3 hour increments on the x-axis
plt.title(f"Electricity Price Changes on {specific_day}")
plt.xlabel("Time")
plt.ylabel("RRP (Regional Reference Price)")
plt.legend(title="Region")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Plot long term price trends based on a monthly average for each state

# Ensure SETTLEMENTDATE is datetime
price_data["SETTLEMENTDATE"] = pd.to_datetime(price_data["SETTLEMENTDATE"])

# Create a Year-Month column for grouping
price_data["YearMonth"] = price_data["SETTLEMENTDATE"].dt.to_period("M")

# Group by Year-Month and REGIONID, then calculate the average RRP
monthly_avg = price_data.groupby(["YearMonth", "REGIONID"])["RRP"].mean().reset_index()

# Convert YearMonth to datetime for plotting
monthly_avg["YearMonth"] = monthly_avg["YearMonth"].dt.to_timestamp()

# Pivot the data to have regions as columns
pivoted_monthly_avg = monthly_avg.pivot(index="YearMonth", columns="REGIONID", values="RRP")

# Plot the monthly averages for each state
plt.figure(figsize=(14, 8))
for region in pivoted_monthly_avg.columns:
    plt.plot(pivoted_monthly_avg.index, pivoted_monthly_avg[region], label=region)

plt.title("Monthly Average Electricity Prices (2009-2024)")
plt.xlabel("Year-Month")
plt.ylabel("Average RRP (Regional Reference Price)")
plt.legend(title="Region")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Group by 'REGIONID' and calculate summary statistics for 'RRP'
summary_stats = price_data.groupby('REGIONID')['RRP'].describe(percentiles=[.25, .5, .75])

# Rename columns for clarity
summary_stats = summary_stats.rename(columns={
    'count': 'Count',
    'mean': 'Mean',
    'std': 'Standard Deviation',
    'min': 'Minimum',
    '25%': '25th Percentile',
    '50%': 'Median',
    '75%': '75th Percentile',
    'max': 'Maximum'
})

# Generate LaTeX code for the summary statistics table
latex_table = summary_stats.to_latex(
    index=True,
    column_format='lrrrrrrrr',  # Adjust column alignment: 'l' for left, 'r' for right
    bold_rows=True,
    caption='Summary Statistics of Regional Reference Prices',
    label='tab:summary_stats'
)


# Display the LaTeX table in the notebook
display(Latex(latex_table))
```
